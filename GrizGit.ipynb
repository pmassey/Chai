```python
#Import various things:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 


```

```python
#Now lets see if we can create a data frame using pandas
grizdata = pd.read_excel("/users/massey/dropbox/Griz/GrizWt.xlsx")
grizdata.describe()
grizdata.plot(kind='scatter', x='DeltaDays', y='oz', title='Griz wt')
```

```python
y=6+grizdata.oz/16.
x=grizdata.DeltaDays
plt.scatter(x,y)
plt.title('Grizs weight as a function of date')
plt.xlabel('Delta Days')
plt.ylabel('Weight in pounds')
```

```python
# --- plotting setup ----
f = plt.figure(facecolor='white', figsize=(3,2), dpi=250)
ax1 = plt.subplot(111)
ax1.scatter(x,y,1)
ax1.set_xlabel('Delta days', size='small', style='normal', family='sans-serif', fontname='Helvetica')
ax1.set_ylabel('Weight in pounds', size='small', style='normal', family='sans-serif', fontname='Helvetica')
fit_coefs = np.polyfit(x,y,2)
print(fit_coefs)
deriv = (2 * fit_coefs[0] * 60 + fit_coefs[1]) * 12 * 7
print(f'Griz is losing {deriv:1.1f} oz/week')
fit_linear = np.poly1d(fit_coefs)
ax1.plot(x,fit_linear(x),'-',lw=1)
```
